<h3>3-2 : 1. 각 경찰서(`관서명`)를 해당 구 이름으로 매핑하여 '구별'이라는 새로운 column을 생성하고, DataFrame을 출력해주세요.<br>
    - 매칭되지 않는 경찰서명에 대해서는 기본값 `'구 없음'`을 할당합니다.</h3>

```python
#각 경찰서와 구를 딕셔너리로 저장해서 매칭할 때 사용
area = {'서대문서': '서대문구', '수서서': '강남구', '강서서': '강서구', '서초서': '서초구',
'서부서': '은평구', '중부서': '중구', '종로서': '종로구', '남대문서': '중구',
'혜화서': '종로구', '용산서': '용산구', '성북서': '성북구', '동대문서': '동대문구',
'마포서': '마포구', '영등포서': '영등포구', '성동서': '성동구', '동작서': '동작구',
'광진서': '광진구', '강북서': '강북구', '금천서': '금천구', '중랑서': '중랑구',
'강남서': '강남구', '관악서': '관악구', '강동서': '강동구', '종암서': '성북구', 
'구로서': '구로구', '양천서': '양천구', '송파서': '송파구', '노원서': '노원구', 
'방배서': '서초구', '은평서': '은평구', '도봉서': '도봉구'}
```

```python
#df_crime 데이터프레임에 '구별'이라는 열 생성
df_crime['구별'] = ''
```

```python
# 관서명을 넣었을 때 소속 구가 있으면 구를, 없으면 '구 없음' 출력
def mapping(office):
    if office in area.keys():
        return area[office]
    else:
        return '구 없음'
```

```python
# 데이터프레임이 총 32행이므로 row는 각 행의 인덱스
for row in range(32):
    #mapping의 return 값을 대입
    df_crime['구별'][row] = mapping(df_crime['관서명'][row])
```

```python
#확인 결과 잘 들어갔다.
df_crime
```

<h3>3-3 : pivot_table 을 사용하여 관서별 데이터를 구별 데이터로 변경하고, 같은 구의 경우에는 sum을 적용하여 더해주세요. (index : 관서명 -> 구별)</h3>

```python
#관서명도 합해지므로 필요없어서 관서명 제외한 df 생성
df_no_office = df_crime.drop(columns = '관서명')
#관서명 없는 데이터프레임에서 '구별'을 인덱스로, 합 적용
area_pivot = pd.pivot_table(df_no_office, index = '구별', aggfunc = 'sum')
print(area_pivot)
```

<h3>3-4 : 구 없음  행은 drop 을 활용하여 삭제해주세요.</h3>

```python
#구 이름이 index이므로, '구 없음'이 인덱스인 경우 행 삭제
area_pivot.drop(index = '구 없음', inplace = True)
area_pivot
```

<h3>3-5 : 각 범죄 별로 검거율을 계산하고, 각 검거율 데이터 column을 DataFrame에 추가해주세요.</h3>

```python
#검거율 = 검거 / 발생 이므로 열 전체 값을 한번에 지정
area_pivot['강간 검거율'] = area_pivot['강간(검거)'] / area_pivot['강간(발생)']
area_pivot['강도 검거율'] = area_pivot['강도(검거)'] / area_pivot['강도(발생)']
area_pivot['살인 검거율'] = area_pivot['살인(검거)'] / area_pivot['살인(발생)']
area_pivot['절도 검거율'] = area_pivot['절도(검거)'] / area_pivot['절도(발생)']
area_pivot['폭력 검거율'] = area_pivot['폭력(검거)'] / area_pivot['폭력(발생)']
area_pivot['검거율'] = area_pivot['소계(검거)'] / area_pivot['소계(발생)']
```

```python
#area_pivot은 인덱스가 이미 '구별'로 되어있으므로 바로 조인.
joined_df = area_pivot.join(population)
print(joined_df)
```
